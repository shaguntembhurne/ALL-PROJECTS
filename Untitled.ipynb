```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### basic graph

```python
arrx = np.arange(0,10)
arry = np.arange(10,20)
#plt.plot(arrx,arry,label="2x", linewidth =2, marker= ".", markersize = 10,color = 'red')
#for the short hand notation fmt = color marker line
plt.plot(arrx,arry,'r.-',label ='2x') 

#resize your graph
#plt.figure(figsize=(2,2))

#line 2
x1 = np.arange(0,4.5,0.5)
plt.plot(x1,x1**2,label = "x squared")


plt.title("BASIC FIRST GRAPH") #font can also be changed fontdict
plt.xlabel("x axis")
plt.ylabel("y axis")

#plt.xticks([2,4,6,8])
#plt.yticks([0,5,10,15,20])

plt.legend()
plt.savefig("mygrapoh.png",dpi = 300)
plt.show()
```

### barchart

```python
value = [2,3,4,5]
labels = ["a","b","c","d"]

plt.figure(figsize=(6,4))
bars = plt.bar(labels,value)
#bars[0].set_hatch('/')

patterns =["O","*","/","|"]
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    



plt.savefig("newimasssssge.png",dpi = 300)

plt.show()
```
